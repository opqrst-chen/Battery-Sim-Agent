```python
import pandas as pd
import numpy as np
import os
```

```python
import matplotlib.pyplot as plt
def plot_metrics(df, exp_id=0, is_save_fig=False, is_show_fig=False, exp_type='single', max_duration=0.0):
    # Assign base colors for each group
    base_colors = {
        "Q": "tab:blue",
        "I": "tab:orange",
        "V": "tab:green",
        "total": "tab:red",
        "": "tab:purple"   # for generic "best_rmse_so_far" without prefix
    }

    metrics_cols = [
        'best_Q_rmse_so_far','best_Q_mape_so_far',
        'best_I_rmse_so_far','best_I_mape_so_far',
        'best_V_rmse_so_far','best_V_mape_so_far',
        'best_total_rmse_so_far','best_total_mape_so_far'
    ]

    # X axis: rounds
    x = df['round'] if 'round' in df.columns else (df.index + 1)

    plt.figure(figsize=(12, 7))

    for col in metrics_cols:
        # Decide group by checking substring
        group = ""
        if "_Q_" in col:
            group = "Q"
        elif "_I_" in col:
            group = "I"
        elif "_V_" in col:
            group = "V"
        elif "_total_" in col:
            group = "total"

        color = base_colors[group]
        
        if "rmse" in col:
            # Darker / more opaque for RMSE
            plt.plot(x, df[col], label=col, color=color, alpha=0.9, linewidth=2)
        else:  # mape
            # Lighter / more transparent for MAPE
            plt.plot(x, df[col], label=col, color=color, alpha=0.5, linewidth=2, linestyle="--")

    plt.xlabel("Round")
    plt.ylabel("LOSS")
    plt.title(f"id {exp_id} Best-so-far Metrics Curves (RMSE vs MAPE)")
    plt.grid(True, alpha=0.3)
    plt.legend(ncol=2)
    plt.tight_layout()
    plt.text(plt.xlim()[1]*0.7, plt.ylim()[1]*0.5, f"Total Duration: {max_duration:.1f}s", fontsize=10, color='gray')
    plt.text(plt.xlim()[1]*0.7, plt.ylim()[1]*0.45, f"Final RMSE: {df['best_total_rmse_so_far'].iloc[-1]:.4f}", fontsize=10, color='gray')
    plt.text(plt.xlim()[1]*0.7, plt.ylim()[1]*0.4, f"Final MAPE: {df['best_total_mape_so_far'].iloc[-1]:.4f}", fontsize=10, color='gray')

    if is_save_fig:
        if not os.path.exists(f"./exp_results/figures/{exp_type}"):
            os.makedirs(f"./exp_results/figures/{exp_type}")
        plt.savefig(f"./exp_results/figures/{exp_type}/exp_{exp_id}_metrics.png")
    if is_show_fig:
        plt.show()
    else:
        plt.close()
```

```python
dir_path = os.path.dirname('./exp_results/')
```

```python
exp_type = 'single'
sub_dir = os.path.join(dir_path, exp_type)
print(sub_dir)
for i in os.listdir(sub_dir):
    # print(i)
    for name in os.listdir(os.path.join(sub_dir, i)):
        # print(name)
        file_path = os.path.join(sub_dir, i, name, 'plot','experiments.csv')
        if os.path.exists(file_path):
            df = pd.read_csv(file_path)
            df['total_rmse'] = df['Q_rmse'] + df['V_rmse'] + df['I_rmse']
            df['total_mape'] = df['Q_mape'] + df['V_mape'] + df['I_mape']
            for col in ['Q_rmse', 'Q_mape', 'I_rmse', 'I_mape',
                'V_rmse', 'V_mape', 'total_rmse', 'total_mape']:
                df[f"best_{col}_so_far"] = df[col].cummin()
    
            df['time'] = pd.to_datetime(df['notes'])
            df['duration_s'] = (df['time'] - df['time'].iloc[0]).dt.total_seconds()
            max_duration = float(df['duration_s'].values[-1])
            plot_metrics(df, exp_id=i, is_save_fig=True,is_show_fig=False, exp_type='single', max_duration=max_duration)
    
```
